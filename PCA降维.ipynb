逻辑回归

```python
'''
检查特征重要性的最简单方法是检查模型的系数。例如，线性回归和逻辑回归都归结为一个方程，其中将系数(重要性)分配给每个输入值。
简单地说，如果分配的系数是一个大(负或正)数字，它会对预测产生一些影响。相反，如果系数为零，则对预测没有任何影响。
逻辑非常简单，让我们来测试一下，逻辑回归是一种合适的算法。拟合模型后，系数将存储在属性中coef_。
该方法最大特点：简单、高效。系数越大(在正方向和负方向)，越影响预测效果。
'''
```

```python
'''
PCA().fit(X) 和 PCA().fit(X_scaled) 的主要区别在于数据预处理的步骤不同。在进行 PCA 之前，通常需要对原始数据进行预处理，
以便在 PCA 中能够得到更好的结果。在这里，特征缩放是一种常见的预处理方法，它将每个特征的值缩放到一个相似的范围内，以确保每个特征对结果的影响相同。因此，如果您在进行 PCA 之前对数据进行了特征缩放，则应该使用 PCA().fit(X_scaled)。如果您没有对数据进行任何特征缩放，则应该使用 PCA().fit(X)。
值得注意的是，使用特征缩放可能会对结果产生重要的影响，因为它可以帮助 PCA 找到更好的主成分，并提高模型的性能。
'''
```

```python
# PCA
from sklearn.decomposition import PCA
pca = PCA().fit(X_train_scaled)  #对特征缩放之后的进行训练
# 拟合数据并获取每个主成分对应的方差解释比例
variance_ratios = pca.explained_variance_ratio_
# 计算累计方差解释比例并绘制图表
cumulative_ratios = np.cumsum(variance_ratios)
# 可视化
#plt.plot(pca.explained_variance_ratio_.cumsum(), lw=3, color='#087E8B')
plt.plot(cumulative_ratios, lw=3, color='#087E8B')
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.title('Cumulative explained variance by number of principal components', size=20)
plt.savefig('/home/aistudio/降维训练/imgs/pca_component.png')
```

```python
# 根据阈值选择主成分个数
threshold = 0.95
n_components = np.argmax(cumulative_ratios >= threshold) + 1
print("Number of components selected:", n_components)
```

```python
# 使用PCA进行降维
pca = PCA(n_components=n_components)
# 将训练数据和测试数据转换为主成分空间   X_train_pca = pca.transform(X_train)
X_transformed = pca.fit_transform(X_train_scaled)
```
